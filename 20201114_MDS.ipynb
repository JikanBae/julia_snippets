# Juliaで多次元尺度法をやってみる
多次元尺度法（MultiDimensional Scaling）のなかでも古典的な方法はMultivariateStatsパッケージに入っている。

https://multivariatestatsjl.readthedocs.io/en/stable/cmds.html

```julia
using MultivariateStats
using Plots
```

```julia
# 渡辺洋 編『心理統計の技法』の「表13-1 野菜の非類似度データ」を使う
D = [
    0.00 7.72 5.94 5.58 6.22 5.70 6.50
    7.72 0.00 5.42 6.24 6.18 6.86 3.26
    5.94 5.42 0.00 4.18 7.44 7.04 4.86
    5.58 6.24 4.18 0.00 6.86 6.38 6.06
    6.22 6.18 7.44 6.86 0.00 4.36 5.10
    5.70 6.86 7.04 6.38 4.36 0.00 6.14
    6.50 3.26 4.86 6.06 5.10 6.14 0.00
];
```

```julia
X = classical_mds(D, 2)'
```

```julia
scatter(-X[:,1], -X[:,2],
    markercolor=:gray, markerstrokecolor=:black,
    legend=false, xlim=(-3.5, +3.5), ylim=(-3.5, +3.5),
    size=(320, 320), dpi=150)
annotate!([
        (-X[1,1], -X[1,2]+0.3, Plots.text("tomato", 7, :center))
        (-X[2,1], -X[2,2]+0.3, Plots.text("daikon", 7, :center))
        (-X[3,1], -X[3,2]+0.3, Plots.text("kyuri", 7, :center))
        (-X[4,1], -X[4,2]+0.3, Plots.text("nasubi", 7, :center))
        (-X[5,1], -X[5,2]+0.3, Plots.text("jagaimo", 7, :center))
        (-X[6,1], -X[6,2]+0.3, Plots.text("kabocha", 7, :center))
        (-X[7,1], -X[7,2]+0.3, Plots.text("ninjin", 7, :center))
        ])
```

```julia
X = classical_mds(D, 1)'
```

```julia
scatter(-X, zeros(length(X)),
    markercolor=:gray, markerstrokecolor=:black,
    legend=false, xlim=(-3.5, +3.5), ylim=(-1, +1),
    size=(320, 320), dpi=150)
annotate!([
        (-X[1,1], 0.1, Plots.text("tomato", 7, :center))
        (-X[2,1], -0.1, Plots.text("daikon", 7, :center))
        (-X[3,1], 0.1, Plots.text("kyuri", 7, :center))
        (-X[4,1], 0.1, Plots.text("nasubi", 7, :center))
        (-X[5,1], -0.1, Plots.text("jagaimo", 7, :center))
        (-X[6,1], 0.1, Plots.text("kabocha", 7, :center))
        (-X[7,1], -0.1, Plots.text("ninjin", 7, :center))
        ])
```
